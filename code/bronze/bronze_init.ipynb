##Extracting and Loading Data from sources into Bronze 

```python
for item in extract_config:
    print(f"Extracting from: {item['source']} into: {item['table_name']}")

    df = spark.read.option('header', True).option('inferSchema', True).csv(item['path'])

    df.write.mode(item['mode']).saveAsTable(item['table_name'])


```
