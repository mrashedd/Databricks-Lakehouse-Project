##Creating View for dim_customers

```python
%sql
CREATE OR REPLACE VIEW gold.dim_customers AS
select
    row_number() over(order by cc.customer_id) as customer_key,
    cc.customer_id,
    cc.customer_number,
    cc.first_name,
    cc.last_name,
    timestampdiff(year, ec.birthdate, current_date()) as age,
    el.country,
    case 
        when lower(cc.gender) = 'unknown' then ec.gender
        else cc.gender
    end as gender,
    cc.marital_status,
    cc.create_date

from silver.crm_customer_info cc
left join silver.erp_customer_az12 ec
on cc.customer_number = ec.customer_number
left join silver.erp_location_a101 el
on cc.customer_number = el.customer_number
```
