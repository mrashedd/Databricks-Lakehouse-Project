##Creating View for dim_products

```python
%sql
CREATE or REPLACE VIEW gold.dim_products AS
select
  ROW_NUMBER() OVER (ORDER BY product_start_date, product_id) AS product_key,
  cp.product_id,
  cp.product_key as product_number,
  cp.product_name,
  ep.category_id,
  ep.category,
  ep.subcategory,
  cp.product_cost,
  cp.product_line,
  ep.maintenance,
  cp.product_start_date as start_date
from silver.crm_product_info cp
left join silver.erp_px_category ep
on cp.product_category_id = ep.category_id
WHERE product_end_date is null
```
